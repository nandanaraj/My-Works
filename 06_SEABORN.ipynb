```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'])
```

```python
tips.info()
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
plt.scatter(tips['total_bill'],tips['size'])
```

```python
plt.scatter(tips['size'],tips['tip'])
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker')
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
tips['total_bill']=tips['total_bill'].astype(np.int32)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
plt.figure(figsize=(15,6))
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
tips.corr() ##Correlation
```

```python
plt.figure(figsize=(18,7))
sns.heatmap(tips.corr(),annot=True)
```

```python
plt.figure(figsize=(18,7))
sns.heatmap(tips.corr(),annot=True,cmap='Pastel1')
```

```python
sns.FacetGrid(data=tips,row='smoker',col='time')
## created a blank canvas using these 2options in smoker and 2 options in time,that makes it 4
```

```python

sns.FacetGrid(data=tips,row='smoker',col='time').map(plt.scatter,'total_bill','tip')
```

```python
sns.catplot(data=tips,x='day',y='tip')
```
